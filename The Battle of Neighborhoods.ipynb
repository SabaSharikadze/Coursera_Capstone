## The Battle of Neighborhoods - Opening Martial Arts Center in Chicago

&ensp;

### Part 1: Introduction/Business Problem

Let us assume that you want to open up a martial arts school somewhere in the city of Chicago and obviously looking for ways to make it profitable and long lasting. There are you and a couple of other practitioners who are willing to put money in this project and out of various issues to solve there is a clear problem that needs to be resolved, namely that of location. Where would it be optimal to base the school? To be able to answer this question there are multiple parameters that need to be considered and many more basic questions to be answered:

- There are hundreds of martial arts studios in Chicago - some of them are long rooted and have other branches in other cities as well while others are smaller and local. So first of all location has to be chosen based on what the competition is in that area. Since this is a new business for you, you would not want to compete with well established centers.


- Customers probably would not like to drive too far from their home especially in Chicago to train hard so it might be wise to choose a location that is close to residential areas and preferably with availability of public transport.


- Assuming limited budget for the starting business, opening up in higher income areas and paying huge rent would probably not be viable even though we could charge customers more.


- Which discipline are we going to concentrate on? Maybe there are some disciplines more popular in some areas of Chicago while others are popular in other areas. We could try clustering disciplines and see if they are lumped in specific locations or scattered all over the map.

Foursquare would help us with answering some of the questions above. We could find all the existing martial arts centers in Chicago and group them in terms of how many there are in each neighborhood or district. With the help of clustering algorithm and some basic data visualization tools we could answer the questions above and get a better idea which location would be ideal for opening up a new martial arts studio.

### Part 2: Data Description

First of all we need to segment the full area of Chicago into well defined areas. We could go with the neighborhoods but the problem is that Chicago does not have well defined neighborhoods. Wikipedia page says that "there are sometimes said to be more than 200 neighborhoods in Chicago, though residents differ on their names and boundaries. A city ordinance prescribing and mapping 178 neighborhoods is almost unknown and ignored even by municipal departments." Another way to go is to segment Chicago in terms of community areas and since there are very well defined 77 community areas it should be better way to go.

The table of community areas can be found on the wikipedia page:
<a href = "https://en.wikipedia.org/wiki/Community_areas_in_Chicago">Community areas in Chicago</a>
 so we can directly extract the table either by beautiful soup library as html text or as a json file. Then using geocoder library we can assign coordinates to the corresponding community areas. Additionally, the same wikipedia page contains not only community areas but also their respective population and area, so that can come in handy as well when deciding where to open up martial arts studio. Foursquare data will be used to explore venues for every community area, namely for every community area we can check how many martial arts schools are available and how long have they been there. We can even look at the ratings and see if there is any correlation at all between community area and average ratings.
 
Finally we can further categorize already existing martial arts schools in Chicago into what disciplines they teach, ex. Taekwondo, Karate, Ju-Jitsu, etc. Then we can cluster the data in terms of prevalence of these disciplines. Maybe in some locations Judo is more popular and prevalent whereas in others Karate is more popular. We can also use another clustering but this time on the school ratings and the duration of their existence.
